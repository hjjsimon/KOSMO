```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Numpy배열 생성하기 4 : 랜덤한 데이타로 배열 생성
#1. normal()함수
'''
정규분포(평균이 0 표준표차가 1인)를 따르는 데이타에서 랜덤하게 값 샘플링:normal
정규 분포 에서 랜덤하게 추출한 샘플값으로 배열 생성
loc는 평균.scale는 표준편차

'''
normal = np.random.normal(loc=0,scale=1,size=1000)
print('평균:',np.mean(normal))
print('표준편차:',np.std(normal))
print(normal.shape)
normal
```

```python
plt.hist(normal)
plt.show()
```

```python
#2. rand()함수
#균등분포를 따르는 확률값에서 랜덤하게 0에서 1사이의 값 샘플링:rand
rand = np.random.rand(5,2)#0과 1의 범위(확률값)의 균등분포에서 랜덤한 값으로 초기화 된 shape가 (5,2)배열 생성
rand
```

```python
plt.hist(rand)
plt.show()
```
