```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''
https://www.kaggle.com/competitions/titanic/data


PassengerId:승객번호
Survived:생존 여부(0 = No, 1 = Yes)-train 데이터에서만 제공
Pclass:좌석 등급(1 = 1st, 2 = 2nd, 3 = 3rd)
Name:이름
Sex:성별
Age:나이
SibSp: 형제 자매 / 배우자 수
Parch:부모와 자식 수(어떤 아이들은 유모와 탑승했으며, 그 결과 0이 있음)
Ticket:티켓 번호
Fare:티켓 요금
Cabin:객실 번호
Embarked:승선한 항구(C = Cherbourg, Q = Queenstown, S = Southampton)
'''
from urllib.request import urlretrieve
```

```python
file_name= 'titanic.csv'
file_id='1T84Ub1ZcY9sE8M5ojMUQUCP03lKTgKji'
url = f'https://drive.google.com/uc?id={file_id}'
urlretrieve(url,file_name)
```

```python
#파일 읽기 첫번째
df= pd.read_csv(file_name)#csv파일을 데이터프레임으로 읽어 들인다
df
```

```python
df.info()
```

```python
df.columns#컬럼들 보기
```

```python
#생존자만 보기:Survived == 1:불리언 마스크 사용
mask = df['Survived']==1##불리언 마스크:query()함수 미 사용
mask
```

```python
df[mask]
```

```python
#생존자만 보기-query()함수 사용
df.query(expr='Survived==1')#Survived==1인 모든 행 추출.SQL의 WHERE절과 같다.
```

```python
#Cabin컬럼에 NaN이 포함된 행들만 추출-불리언 마스크 사용
df[df['Cabin'].isna()]
```

```python
#Cabin컬럼에 NaN이 포함된 행들만 추출-query()함수 사용
df.query(expr='Cabin.isna()')
```

```python
#파일 읽기 두번째
df= pd.read_csv(file_name,skiprows=[0],names=['승개번호', '생사여부', '좌석등급', '이름', '성별', '나이', '형제/자매/베우자수','부모/자식수','티켓번호', '티켓요금', '객실번호', '승선항수'])
df
```

```python
df.info()
```

```python
df['성별'].sample(10)
```

```python
#성별이 female인 탐승객들 추출-불리언 마스크 사용
df[df['성별']=='female'][['성별','나이']]
```

```python
#성별이 female인 탐승객들 추출-query()함수 사용
df.query('성별=="female"')[['성별','나이']]
```

```python
#파일 읽기 세번째
#index_col는 특정 컬럼의 값을 행인덱스로 사용
#index_col는 모든행이 NaN아니고 유니크해야 한다
df= pd.read_csv(file_name,index_col='PassengerId',sep=',')#sep는 디폴트가 ,(콤마) csv파일이니까 (탭으로 구분된 csv도 있다:.tsv라고 확장자)
df.head()
```

```python
#파일 읽기 네번째
#탭으로 구분되 있지 않고 빈 공백으로 구분되 있다.(빈 공백수가 다르다)
#\s빈 공백이고 +는 하나이상
df = pd.read_csv('blank.csv',sep='\s+',names=['이름','나이','성별','거주지'])
df
```

```python
#파일 읽기 다섯번째
#na_values인수로 특정값을 NaN 처리하기
df = pd.read_csv('blank.csv',sep='\s+',names=['이름','나이','성별','거주지'],na_values=['나길동',35])
df
```

```python
#데이타 프레임을 파일로 저장하기:to_csv()함수
df.to_csv('blank_new1.csv')#행 인덱스 포함(디폴트)
df.to_csv('blank_new2.csv',index=False)#행 인덱스 제외
df.to_csv('blank_new3.csv',index=False,sep='|')#sep인수로 파일로 저장시 구분자 바꾸기
df.to_csv('blank_new4.csv',index=False,sep='|',na_rep='결측치')#na_rep인수로 NaN값을 특정 값으로 바꾸기
df.to_csv('blank_new5.csv',index=False,sep='\t',na_rep='결측치',header=False)#header=False로 파일저장시 컬럼인덱스 제외
```
