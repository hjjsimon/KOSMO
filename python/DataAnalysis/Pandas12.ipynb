```python
import numpy as np
import pandas as pd
```

```python
#데이타 그룹화하기
np.random.seed(0)
df = pd.DataFrame({'A':['foo','bar','foo','bar','foo','bar','foo','foo'],
                   'B':['one','one','two','three','two','two','one','three'],
                   'C':np.random.randint(1,8,8),
                   'D':np.random.randint(1,8,8)})
df
```

```python
#단일 함수(집합함수 혹은 그룹함수)를 사용하여 GroupBy 집계
#1.컬럼 색인으로 그룹핑
gby=df.groupby('A')#<pandas.core.groupby.generic.DataFrameGroupBy object at 0x0000026AE35F2710>
dir(gby)
```

```python
gby.sum()#foo그룹 및 bar그룹의 각 컬럼의 평균을 구한다
         #select sum('C'),sum('D') from 테이블명 group by 'A'
```

```python
gby.count()#각 컬럼별 bar와 foo의 행의 수
```

```python
gby.max()#각 컬럼별 최대값
```

```python
gby.agg('max')##혹은 DataFrameGroupBy객체.agg('함수명')
```

```python
#보통은 아래처럼 메소드체인 사용
print(df)
df.groupby(by='A',sort=False).max()#그룹핑한 컬럼명을 정렬(groupby함수에 sort인자가 있다:sort=True 디폴트. 그룹키(by지정한 컬럼명)로 오름차순)

df.groupby(by='A').max().sort_index(ascending=False)#그룹핑한 컬럼명을 정렬.디폴트는 오름차순이고 False지정시 내림차순
df.groupby(by='A',sort=False).agg('min')
```

```python
#여러 컬럼을 기준으로 그룹핑하여 집계
#select min(C),min(D) from emp group by A,B
df.groupby(['A','B']).min()
```

```python
#DataFrameGroupBy객체.agg(["집계함수1","집계함수2"...])
##A와 B컬럼으로 그룹핑해서 C와 D컬럼에 대한 기술통값 구하기
df.groupby(['A','B']).agg(['sum','mean','min','max','size','count'])
```

```python
#agg({'컬럼명1':['집계함수1','집계함수2'...],'컬럼명2':['집계함수1','집계함수2'...]})
df.groupby(['A','B']).agg({'C':['sum','mean'],'D':['min','max']})
```

```python
#('별칭명','함수명')
df.groupby(['A','B']).agg({'C':[('총합','sum'),('평균','mean')],'D':[('최소값','min'),('최대값','max')]})
```
