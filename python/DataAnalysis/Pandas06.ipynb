```python
import numpy as np
import pandas as pd
```

```python
#rand: 주어진 shape으로 0에서 1사이의 균등분포를 따른 값을 무작위로 샘플링
df = pd.DataFrame(np.random.rand(5,5))
df
df.index#RangeIndex(start=0, stop=5, step=1)
df.columns#RangeIndex(start=0, stop=5, step=1)
pd.date_range('20230818',periods=5)#DatetimeIndex(['2023-08-18', '2023-08-19', '2023-08-20', '2023-08-21', '2023-08-22'], dtype='datetime64[ns]', freq='D')
df.index = pd.date_range('20230818',periods=5) 
df.columns =list('가나다라마')  
df
```

```python
#1.dropna() 함수
'''
dropna()함수: 결측치가 존재하는 행을 제외
how="any" : 행의 각 컬럼에 NaN이 하나라도 있으면 그 행을 제거(디폴트)
how="all" : 행의 모든 컬럼이 모두 NaN일때 그 행을 제거
axis=0 행 제거(디폴트), 열 제거시에는 axis=1
inplace =True 일때는 반환값이 없다(왜냐하면 원본이 변하니까)
'''
df['바']=0.5,np.nan,1.0,np.nan,100
df
```

```python
#df_=df.dropna(how='any',axis=0)#df.dropna() 컬럼에 하나라도 NaN이 있는 행은 제거됨
#df_=df.dropna(how='all')#모든 컬럼에 NaN이 있는 행이 제거됨
#df_ = df.dropna(axis=1)#NaN이 하나라도 있는 열이 제거됨
df_=df.dropna(how='all',axis=1)#모든 행에 NaN이 있는 컬럼이 제거됨
df_
```

```python
#2.fillna() 함수
#결측치가 있는 부분을 다른 값으로 채우는 함수
df.fillna(0.1)
```

```python
#3.isna() 함수 혹은 isnull()함수
#결측치에 해당하면 True,아니면 False로 불린 마스크를 생성하는 함수
#df.isna()
df.isnull()
df.isnull().sum()#각 컬럼별 결측피 수(시리즈)
df.isnull().to_numpy().sum()#스칼라 2
```

```python
#불리언 마스크를 사용해서 바열에서 NaN인 행만 모두 가져오기
df[df.isnull()['바']]#혹은 df.loc[df.isnull()['바']]
```

```python
#'바'열에서 100이 포함된 행을 제거하시오
df < 100
```

```python
df_ = df#원본 저장
df_.fillna(0.5,inplace=True)#NaN을 임의값으로 설정
mask = (df_ < 100)['바']#마스크 생성(100이 존재하는 행이 False가 되록 조건 작성)
mask
df = df.loc[mask,:]#마스크를 DF에 적용. 그러면 100이 있는 행은 다 제거 된다
df
```

```python
del df['바']#바 열 하나 삭제.원본 데이타프레임이 변경됨.여러 컬럼 삭제는 지원하지 않는다 대신 drop()함수 사용
```

```python
#여러열 삭제시는 drop()함수
df.drop(['가','다'],axis=1)#not inplcae
df
```
